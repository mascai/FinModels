```r
ret_foo <- function(data) { # returns stock return: return_t = (price_t - price_t-1) / price_t-1
    
    v2 = data[-1]
    v2 <-append(v2, 0)
    ret = (data - v2) / v2

    print(data)
    print(v2)
    
    ret[-length(ret)]
}

```
